```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train=pd.read_csv('train.csv')
train
```

```python
test=pd.read_csv('test.csv')
test
```

```python
from sklearn.preprocessing import LabelEncoder
#for train data

le_th=LabelEncoder() #for home ownership
le_ti=LabelEncoder() #for loan intent
le_tg=LabelEncoder() #for loan grade
le_td=LabelEncoder() #for person on default file
train['owner']=le_th.fit_transform(train['person_home_ownership'])
train['intent']=le_ti.fit_transform(train['loan_intent'])
train['grade']=le_tg.fit_transform(train['loan_grade'])
train['default']=le_td.fit_transform(train['cb_person_default_on_file'])

#for test data
le_tth=LabelEncoder() #for home ownership
le_tti=LabelEncoder() #for loan intent
le_ttg=LabelEncoder() #for loan grade
le_ttd=LabelEncoder() #for person on default file
test['owner']=le_tth.fit_transform(test['person_home_ownership'])
test['intent']=le_tti.fit_transform(test['loan_intent'])
test['grade']=le_ttg.fit_transform(test['loan_grade'])
test['default']=le_ttd.fit_transform(test['cb_person_default_on_file'])
```

```python
#Dropping columns not required anymore
target=train['loan_status']
train=train.drop(['person_home_ownership','loan_intent','loan_grade','cb_person_default_on_file','loan_status'],axis='columns')
test=test.drop(['person_home_ownership','loan_intent','loan_grade','cb_person_default_on_file'],axis='columns')
```

```python
#Findng any NA values
print('TRAIN: \n',train.isna().any())
print('\n TEST: \n',test.isna().any())
```

```python
#Finding model through hyperparametre tuning
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

model_params={
    'RandomForest':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators': [2,4,6,8]
        }
    },

    'LogisticRegression':{
        'model': LogisticRegression(solver='liblinear'),
        'params':{
            'C': [2,4,6,8]
        }
    },

    'GaussianNB':{
       'model': GaussianNB(),
        'params':{
            'var_smoothing': np.logspace(0,-9, num=100)

        }
    },

    'MultinomialNB':{
        'model': MultinomialNB(),
        'params':{

        }
    },

    'DecisionTreeClassifier':{'model': DecisionTreeClassifier(),
        'params':{
            'criterion': ['gini','entropy']
        }
    }


}
```

```python
from sklearn.model_selection import GridSearchCV
scores=[]
for mn,mp in model_params.items():
  print(mn)
  print(mp)
  clf= GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
  clf.fit(train,target)
  scores.append({
      'model':mn,
      'best_params':clf.best_params_,
      'best_score':clf.best_score_
  })
  df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
model=RandomForestClassifier(n_estimators=2)
model.fit(train,target)
model.score(train,target)
```

```python
y_pred= model.predict(test)
```

```python
final=pd.DataFrame()
final['id']=test['id']
final['loan_status']=y_pred
final
```

```python
final.to_csv('loan_approval.csv')
```
